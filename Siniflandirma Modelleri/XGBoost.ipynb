# XGBoost

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
data_file = "diabetes.csv"
```

```python
df = pd.read_csv(data_file)
df.head()
```

```python
X = df.drop(["Outcome"], axis=1)
y = df["Outcome"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3232)
```

```python
xgb_model = XGBClassifier().fit(X_train, y_train)
```

# Tahmin

```python
y_pred = xgb_model.predict(X_test)
accuracy_score(y_test, y_pred)
```

# Model Tuning

```python
xgb_params = {'n_estimators': [100, 250, 500, 2100],
              'subsample': [0.8, 0.9, 1.0],
              'max_depth': [3, 4, 5, 6],
              'learning_rate': [0.01, 0.1, 0.2],
              'min_samples_split': [3, 7, 10]}
xgb_model = XGBClassifier()
xgb_cv_model = GridSearchCV(xgb_model, xgb_params, cv=10, n_jobs=-1).fit(X_train, y_train)
```

```python
xgb_cv_model.best_params_
```

```python
xgb_tuned = XGBClassifier(learning_rate=0.01, max_depth=3, min_samples_split=3, n_estimators=250, subsample=0.8).fit(X_train, y_train)
y_pred = xgb_tuned.predict(X_test)
accuracy_score(y_test, y_pred)
```
