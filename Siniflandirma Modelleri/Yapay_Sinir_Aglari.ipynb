# Yapay Sinir Aglari

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
data_file = "diabetes.csv"
```

```python
df = pd.read_csv(data_file)
df.head()
```

```python
X = df.drop(["Outcome"], axis=1)
y = df["Outcome"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3232)
```

```python
scaler = StandardScaler().fit(X_train, y_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
mlpc_model = MLPClassifier().fit(X_train_scaled, y_train)
```

# Tahmin

```python
y_pred = mlpc_model.predict(X_test)
accuracy_score(y_test, y_pred)
```

# Model Tuning

```python
mlpc_params = {"alpha": [0.0001, 0.001, 0.01, 0.1],
               "hidden_layer_sizes": [(10, 10, 10), (100, 100, 100)],
               "solver": ["lbfgs", "adam", "sgd"],
               "activation": ["relu", "logistic"]}
mlpc = MLPClassifier()
mlpc_cv_model = GridSearchCV(mlpc, mlpc_params, cv=10, n_jobs=-1).fit(X_train, y_train)
```

```python
mlpc_cv_model.best_params_
```

```python
mlpc_tuned = MLPClassifier(activation="logistic", alpha=0.1, hidden_layer_sizes=(100, 100, 100), solver="adam").fit(X_train, y_train)
y_pred = mlpc_tuned.predict(X_test)
accuracy_score(y_test, y_pred)
```
