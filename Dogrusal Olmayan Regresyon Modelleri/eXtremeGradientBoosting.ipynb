# eXtreme Gradient Boosting (XGBoost)

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
```

```python
data_file = "Hitters.csv"
```

```python
df = pd.read_csv(data_file).dropna()
df.head()
```

```python
x64 = df.drop(["Salary", "League", "Division", "NewLeague"], axis=1).astype("float64")
y = df["Salary"]
dummies = pd.get_dummies(df[["League", "Division", "NewLeague"]])
X = pd.concat([x64, dummies[["League_N", "Division_W", "NewLeague_N"]]], axis=1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3232)
DM_train = xgb.DMatrix(data=X_train, label=y_train)
DM_test = xgb.DMatrix(data=X_test, label=X_test)

xgb_model = XGBRegressor().fit(X_train, y_train)
```

# Tahmin

```python
y_pred = xgb_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

# Model Tuning

```python
xgb_params = {'colsample_byree': [0.5, 0.6, 0.7, 0.8, 0.9, 1],
            'n_estimators': [100, 250, 500, 1000],
            'max_depth': [2, 3, 4, 5],
            'learning_rate': [0.01, 0.5, 0.1]}
xgb_cv_model = GridSearchCV(xgb_model, xgb_params, cv=10, n_jobs=-1).fit(X_train, y_train)
xgb_cv_model.best_params_
```

```python
xgb_tuned = XGBRegressor(colsample_byree=0.5, learning_rate=0.1, max_depth=2, n_estimators=250).fit(X_train, y_train)
y_pred = xgb_tuned.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```
