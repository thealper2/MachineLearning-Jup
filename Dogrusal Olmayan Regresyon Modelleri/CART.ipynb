# CART

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
data_file = "Hitters.csv"
```

```python
df = pd.read_csv(data_file).dropna()
df.head()
```

```python
x64 = df.drop(["Salary", "League", "Division", "NewLeague"], axis=1).astype("float64")
y = df["Salary"]
dummies = pd.get_dummies(df[["League", "Division", "NewLeague"]])
X = pd.concat([x64, dummies[["League_N", "Division_W", "NewLeague_N"]]], axis=1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3232)
X_train = pd.DataFrame(X_train["Hits"])
X_test = pd.DataFrame(X_test["Hits"])
```

```python
cart_model = DecisionTreeRegressor(max_leaf_nodes=10).fit(X_train, y_train)
```

# Tahmin

```python
y_pred = cart_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

# Model Tuning

```python
cart_params = {'min_samples_split': range(2, 100), 'max_leaf_nodes': range(2, 10)}
cart_cv_model = GridSearchCV(cart_model, cart_params, cv=10).fit(X_train, y_train)
cart_cv_model.best_params_
```

```python
cart_tuned = DecisionTreeRegressor(min_samples_split=2, max_leaf_nodes=2).fit(X_train, y_train)
y_pred = cart_tuned.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```
